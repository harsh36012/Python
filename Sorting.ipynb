# Sorting 

```python
# check an array for sort

def is_sort(arr):
    for i in range(len(arr)-1):
        if arr[i] > arr[i+1]: return False
    return True
```

```python
arr = [4,11,15,29,40,56,60,71,84,95,100]
is_sort(arr)
```

### Monkey Sort

```python
# it is totally on random guess
# Time Complexity -> Best case - O(1) , Worst case - O(infinite)
# Space Complexity -> O(1)
```

```python
# monkey sort
import random

def monkey_sort(arr):
    while not is_sort(arr):
        random.shuffle(arr)
        print(arr)
```

```python
arr = [7,4,2,10]
```

```python
monkey_sort(arr)
```

### Bubble Sort

```python
# Time Complexity -> Best Case - O(n^2) , Worst Case - O(n^2)
# Space Complexity -> O(1)
# Is it Adaptive ?  -> By Nature it is not, but can make it.
# Is it Stable ?   -> Yes 
```

```python
def bubble_sort(arr):
    for i in range(len(arr)-1):  # this is for the passes
        for j in range(len(arr) - 1 - i):
            if arr[j] > arr[j+1]:
                # swap
                arr[j],arr[j+1] = arr[j+1],arr[j]

    return arr
```

```python
arr = [17,14,12,10]
bubble_sort(arr)
```

```python
# Adaptivity -> We can make it.
# Time Complexity -> Best Case - O(n) , Worst Case - O(n^2)
```

```python
def bubble_sort(arr):
    for i in range(len(arr)-1):  # this is for the passes
        flag = 0
        for j in range(len(arr) - 1 - i):
            if arr[j] > arr[j+1]:
                # swap
                arr[j],arr[j+1] = arr[j+1],arr[j]
                flag = 1

        if flag == 0:
            # already sorted
            break
    return arr
```

```python
arr = [5,11,17,20]
bubble_sort(arr)
```

### Selection Sort

```python
# select one item and put it in the correct place , and take the next one then put in the 2nd place .and keep on doing this.
# Time Complexity -> Best Case - O(n^2) , Worst Case - O(n^2)
# Space Complexity -> O(1)
# Is it Adaptive ?  -> No
# Is it Stable ?   -> No 
```

```python
def selection_sort(arr):
    for i in range(len(arr)-1):
        Max = i
        for j in range(i+1,len(arr)):
            if arr[j] < arr[Max]:
                Max = j  # new max
        # swap max and i
        arr[Max],arr[i] = arr[i],arr[Max]
    return arr
```

```python
arr = [51,11,97,25,7,5,55]
selection_sort(arr)
```

### Merge Sort

```python
# Time Complexity -> Best & Worst - O(NlogN)
# Space Complexity -> O(n) + O(log N) = O(N)
# adaptive -> no
# stable -> yes
```

```python
def sort_two_sorted_array(arr1,arr2):
    i,j = 0,0
    l = []
    while ((i < len(arr1)) and (j < len(arr2))):
        if arr1[i] < arr2[j]:
            l.append(arr1[i])
            i += 1
        else :
            l.append(arr2[j])
            j += 1

    while i < len(arr1):
        l.append(arr1[i])
        i += 1
    while j < len(arr2):
        l.append(arr2[j])
        j += 1
    return l

def merge_sort(arr):
    if len(arr) == 1:
        return arr
    mid = len(arr) // 2

    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    return sort_two_sorted_array(left,right)

    
```

```python
arr = [51,11,97,25,7,5,55]
print(merge_sort(arr))
```

```python
# More optimized due to in place swapping
```

```python
def sort_two_sorted_array(arr1,arr2,arr):
    i = j = k = 0
    while ((i < len(arr1)) and (j < len(arr2))):
        if arr1[i] < arr2[j]:
            arr[k] = arr1[i]
            i += 1
        else :
            arr[k] = arr2[j]
            j += 1
        k += 1
        
    while i < len(arr1):
        arr[k] = arr1[i]
        i += 1
        k += 1
    while j < len(arr2):
        arr[k] = arr2[j]
        j += 1
        k += 1
    return

def merge_sort(arr):
    if len(arr) == 1:
        return arr
    mid = len(arr) // 2

    left = arr[:mid]
    right = arr[mid:]
    
    merge_sort(left)
    merge_sort(right)

    sort_two_sorted_array(left,right,arr)

    
```

```python
arr = [51,11,97,25,7,5,55]
print(arr)
merge_sort(arr)
print(arr)
```

### Quick Sort

```python
# In this, we use a pivit position and on the bases of it, we dived , and 
# Make sure that pivit value is in its correct position, and left to the pivit are all less , and right are all greater .
# Time Complexity -> Best - O(nlogn) , Worst - O(n^2)
# Space Complexity -> Best - O(logn) , Worst - O(n)
# Adaptive -> No, in general
# stable -> No
```

```python
def partition(arr,st,end):
    idx = st - 1
    pivit = end
    for j in range(st,end):
        if arr[j] < arr[pivit]:
            idx += 1
            arr[j],arr[idx] = arr[idx],arr[j]

    idx += 1
    arr[pivit],arr[idx] = arr[idx],arr[pivit]
    return idx

def Quick_sort(arr,start,end):
    if start < end:
        pivit = partition(arr,start,end)
        Quick_sort(arr,start,pivit-1)  # left 
        Quick_sort(arr,pivit+1,end)    # right
    return 
```

```python
arr = [51,11,97,25,7,5,55]
Quick_sort(arr,0,len(arr)-1)
print(arr)
```

### Insertion Sort

```python
# Time Complexity -> Best Case - O(n) and Worst - O(n^2)
# Space Complexity -> O(1)
# Adaptive -> yes
# stqble -> yes
```

```python
def Insertion_sort(arr):
    for i in range(1,len(arr)):
        curr = arr[i]
        prev = i - 1
        while(prev >= 0 and curr < arr[prev]):
            arr[prev+1] = arr[prev]
            prev -= 1
        arr[prev+1] = curr
    return arr
```

```python
arr = [51,11,97,25,7,5,55]
# arr = [5, 7, 11, 25, 51, 55, 97]
Insertion_sort(arr)
```
