```python
class Node:
    def __init__(self,value):
        self.data = value
        self.next = None

class Stack:
    def __init__(self):
        self.head = None
        self.n = 0

    def __str__(self):
        curr = self.head
        res = ""
        while curr != None:
            res += str(curr.data) + " "
            curr = curr.next
        return res

    

    '''' push operation -> push from head ,top is head '''
    def push(self,value):
        new_node = Node(value)

        new_node.next = self.head
        self.head = new_node
        self.n += 1

    ''' pop -> remove the top value'''
    def pop(self):
        if self.is_empty():
            return "Stack Empty"

        temp = self.head.data
        self.head = self.head.next
        return temp
    
    
    ''' peak -> give the top value'''
    def peak(self):
        if self.is_empty():
            return "Stack Empty"

        return self.head.data
        
    
    ''' isempty '''
    def is_empty(self):
        return self.head == None

    
    ''' size '''
    def size(self):
        return self.n
```

```python
a = Stack()
b = Stack()

# inserting a value
def enqueue(value):
    a.push(value)


# delete a value
def dequeue():
    if not b.is_empty():
        # b has elements 
        b.pop()
    else:
        # b is empty
        # push all the elements in the b stack
        while not a.is_empty():
            data = a.pop()
            b.push(data)
            
        # pushed all elements of a in b in reverse order
        # Now delete from b stack
        b.pop()

```

```python
enqueue(1)
enqueue(2)
enqueue(3)
```

```python
dequeue()
```
