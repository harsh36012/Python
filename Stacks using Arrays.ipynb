```python
class Stack:
    def __init__(self,size):
        self.size = size
        self.__stack = [None] * size
        self.top = -1

    def __str__(self):
        for i in range(self.top + 1):
            print(self.__stack[i], end = ' ') 
        return ""

    def push(self, value):
        # check for overflow
        if self.is_full():
            return "Stack Overflow"

        self.top += 1
        self.__stack[self.top] = value

    def pop(self):
        # check for empty 
        if self.is_empty():
            return "Stack Underflow"

        temp = self.__stack[self.top]
        self.top -= 1
        return temp

    def is_empty(self):
        return self.top == -1

    def is_full(self):
        return self.top == self.size-1
```

```python
s = Stack(3)

s.push(3)
s.push(2)
s.push(1)
```
