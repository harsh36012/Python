```python
class Node:
    def __init__(self,value):
        self.data = value
        self.next = None

class Queue:
    def __init__(self):
        self.front = None  # pop from front
        self.rear = None   # insert from rear
        self.n = 0         # size of the Queue (No. of Nodes)

    ''' insert new value to the Queue'''
    def enqueue(self, value):
        new_node = Node(value)

        # Queue is Empty
        if self.front == None :
            # Both pointed to the new_node
            self.front = new_node
            self.rear = new_node
            self.n += 1  # increase the size
            return

        # Already have element in the Queue.
        # We have to insert from the rear
        self.rear.next = new_node
        self.rear = new_node
        self.n += 1


    
    ''' Delete element from the front''' 
    def dequeue(self):
        if self.is_empty():
            return "Empty Queue"

        self.front = self.front.next
        self.n -= 1

    
    ''' check for empty Queue'''
    def is_empty(self):
        return self.front == None


    ''' print the Queue'''
    def __str__(self):
        # check empty
        if self.is_empty():
            return "Empty Queue"

        temp = self.front
        while(temp != None):
            print(temp.data, end = ' ')
            temp = temp.next
        return ""


    ''' Front item '''
    def front_item(self):
        if self.is_empty():
            return "Empty Queue"
        return self.front.data


    ''' Rear item '''
    def rear_item(self):
        if self.is_empty():
            return "Empty Queue"
        return self.rear.data
```

```python
q = Queue()
```

```python
q.is_empty()
```

```python
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
```
