```python
# Implement Hashing in the python dictionary

# 2 arrays 1. 1 for keys and another for values
# find the hash value of the key ,then put in that location (key index and value index is same)
# 
```

```python
class Dictionary:

    def __init__(self,capacity = 1):
        self.slots = [None] * capacity  # keys array
        self.data = [None] * capacity   # values array
        self.capacity = capacity        # capacity of the array
        self.size = 0                      # size of the array

    ''' insert item into the dictionary'''
    def put_item(self,key,value):

         # check for array size full.
        if self.capacity == self.size:
            # array size is full, we have to increase the capacity 
            self.double_capacity()   # now bigger capacity array.
            
                    
        # find the hash value of the key
        hash_value = self.get_hash_value(key)   # returns a index

        # check for empty space at that index
        if self.slots[hash_value] == None:
            self.slots[hash_value] = key
            self.data[hash_value] = value
            self.size += 1

        else :   # at that index , already has occupied 

            # check that occupied space is key itself or not
            if self.slots[hash_value] == key:
                # key is already present, so update the value.
                self.data[hash_value] = value

            else :
                
                # rehash
                new_hash_value = self.rehash(hash_value)
    
                # check that
                # if new_hash_value space is empty 
                # or new_hash_value space's key is key itself.
                if self.slots[new_hash_value] != None and self.slots[new_hash_value] != key:
                    # both true, rehash
                    new_hash_value = self.rehash(new_hash_value)

                # check if loop ends with None
                if self.slots[new_hash_value] == None:
                    # index space is empty
                    self.slots[new_hash_value] = key
                    self.data[new_hash_value] = value
                    self.size += 1

                # loop ends with ,it find the key 
                else :  
                    # key already persent, so update the value
                    self.data[new_hash_value] = value

                
    ''' can also be able set item with dictionary syntax'''
    def __setitem__(self,key,value):
        return self.put_item(key,value)

        

    # print the dictionary
    def __str__(self): 
        dic = "{"   # opening brackets
        flag = False  # no element 
        for i in range(self.capacity):
            if self.slots[i] != None:
                flag = True  # elememt exist
                dic += str(self.slots[i]) + ": " + str(self.data[i]) + ", "   # add element to dic

        if flag:  # element exist 
            dic = dic[:-2]   # remove 1 extra -> ", "
        dic += "}"  # closing brackets
        return dic


    ''' get the element using key '''
    def get_item(self,key):
        if self.size == 0:  # if size is 0,and try to fetch data 
            return "No element"

        for i in range(self.capacity):
            if self.slots[i] == key:  # element found 
                return self.data[i]

        return "key Not found"
                

    ''' can also be excess with dictionary syntax'''
    def __getitem__(self,key):
        return self.get_item(key)
    
    
    ''' delete an element from the dictionary'''
    def del_item(self,key):

        # serach for the key
        for i in range(self.capacity):
            if self.slots[i] == key:
                self.size -= 1
                self.slots[i] = None
                self.data[i] = None
                return

        # not such key present in the dictionary
        print("key not found")
        return



    ''' delete element from magic method'''
    def __delitem__(self,key):
        return self.del_item(key)

    
    ''' double the capacity'''
    def double_capacity(self):
        
        old_size = self.size   # store old size
        
        # copy all element from the slots and data , and copy in another variable
        temp_slots = self.slots.copy()
        temp_data = self.data.copy()

        self.capacity = self.capacity * 2     # increase the capacity by double
        self.slots = [None] * self.capacity   # also update the size of 2 arrays
        self.data = [None] * self.capacity
        self.size = 0  
        
        # after increase the capacity, put all the element with 
        # calcuate the hash value for new capacity arrays.

        for i in range(old_size):

            # copy back the key and values
            self.put_item(temp_slots[i],temp_data[i])
            


    # hash value is build-in function which return a unique hash value of any non mutable datatype object,
    # sometime it returns neg hash value.
    ''' Find the hash value'''
    def get_hash_value(self,key):
        return abs(hash(key)) % self.capacity   # return index position



    ''' Rehash value with linear probing'''
    def rehash(self,old_hash_value):
        return (old_hash_value + 1) % self.capacity   # return new index position -> next to the previous index


        
```
