```python
'''
Stack -> Linear Data Structure which uses LIFO / FILO
operations ->
1. push
2. pop
3. peak
4. isempty
5. size
'''


''' Stack Using LinkedList'''

class Node:
    def __init__(self,value):
        self.data = value
        self.next = None

class Stack:
    def __init__(self):
        self.head = None
        self.n = 0

    def __str__(self):
        curr = self.head
        res = ""
        while curr != None:
            res += str(curr.data) + "->"
            curr = curr.next
        return res[:-2]

    

    '''' push operation -> push from head ,top is head '''
    def push(self,value):
        new_node = Node(value)

        new_node.next = self.head
        self.head = new_node
        self.n += 1

    ''' pop -> remove the top value'''
    def pop(self):
        if self.is_empty():
            return "Stack Empty"

        temp = self.head.data
        self.head = self.head.next
        return temp
    
    
    ''' peak -> give the top value'''
    def peak(self):
        if self.is_empty():
            return "Stack Empty"

        return self.head.data
        
    
    ''' isempty '''
    def is_empty(self):
        return self.head == None

    
    ''' size '''
    def size(self):
        return self.n


```

```python
l = Stack()
l.push(1)
l.push(2)
l.push(3)
l.push(4)
l.push(9)
```
