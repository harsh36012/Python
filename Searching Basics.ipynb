### Linear Search

```python
# It is a brute ful searching algorithm.
# Not required for a sorted array
# Time Complexity -> Best case - O(1) , Worst case - O(N)
# Space Complexity -> O(1)
```

```python
def linear_search(array,item):
    for i in range(len(array)):
        if(array[i] == item):
            # item found
            return i   # return index
    return -1   # element not found
```

```python
arr = [5,2,19,4,7,20,1]   # array 
element = 1               # element to is be search
linear_search(arr,element)
```

```python
linear_search(arr,5)
```

```python
linear_search(arr,19)
```

```python
linear_search(arr,100)
```

### Binary Search

```python
# Need a sorted array
# Time Complexity -> Best case - O(log n) , Worst case - O(log n)
# Space Complexity -> O(1)
```

```python
# # ''' using normal way '''

# def binary_search(arr,start_index,end_index,item):
#     while start_index <= end_index:
#         mid_index = (start_index + end_index) // 2   # mid index
#         if arr[mid_index] == item:   # check mid_index is item 
#             return mid_index   # element found , and returning index position
#         if arr[mid_index] > item :   # so we will find in left
#             end_index = mid_index - 1    # update end index
#         else :   # so we will find in right
#             start_index = mid_index + 1   # update start index

#     return -1   # element not found


# ''' using Recursion '''

def binary_search(arr,start_index,end_index,item):
    if start_index <= end_index:
        mid_index = (start_index + end_index) // 2 
        if arr[mid_index] == item:
            return mid_index
        if arr[mid_index] > item :  # we will search in left
            return binary_search(arr,start_index,mid_index-1,item)
        else:   # we will search in right
            return binary_search(arr,mid_index+1,end_index,item)
    else:
        return -1  # element not found
            
```

```python
arr = [1,5,9,14,20,53,67,82,100]
binary_search(arr,0,len(arr)-1,1)
```

```python
binary_search(arr,0,len(arr)-1,100)
```

```python
binary_search(arr,0,len(arr)-1,50)
```
