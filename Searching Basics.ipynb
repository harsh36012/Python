```python
# Linear Search 
# It is a brute ful searching algorithm.
# Not required for a sorted array
# Time Complexity -> Best case - O(1) , Worst case - O(N)
# Space Complexity -> O(1)
```

```python
def linear_search(array,item):
    for i in range(len(array)):
        if(array[i] == item):
            # item found
            return i   # return index
    return -1   # element not found
```

```python
arr = [5,2,19,4,7,20,1]   # array 
element = 1               # element to is be search
linear_search(arr,element)
```

```python
linear_search(arr,5)
```

```python
linear_search(arr,19)
```

```python
linear_search(arr,100)
```
